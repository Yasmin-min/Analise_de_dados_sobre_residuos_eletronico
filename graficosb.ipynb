```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = {
    'Junho 24': [42550, 73411], 'Julho 24': [63418, 82853], 'Agosto 24': [105616, 57753],
    'Setembro 24': [48850, 26694], 'Outubro 24': [24921, 39668], 'Novembro 24': [33217, 24970],
    'Dezembro 24': [22420, 31451], 'Janeiro 25': [73411, 9894], 'Fevereiro 25': [27080, 24468],
    'Março 25': [28614, 21396]
}
df = pd.DataFrame(data=d)
df.index.names = ['Entrada 0, Saida 1']
df
```

```python
entradas = df.loc[0]
entradas.plot(kind='bar', color='orange', edgecolor='black')

plt.title('Entradas por Mês')
plt.ylabel('Valor de Entrada')
plt.xlabel('Mês')
plt.xticks(rotation=45, ha='right', color = 'black')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
saidas = df.loc[1]
saidas.plot(kind='bar', color='orange', edgecolor='black')

plt.title('Saídas por Mês')
plt.ylabel('Valor de Saída')
plt.xlabel('Mês')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
